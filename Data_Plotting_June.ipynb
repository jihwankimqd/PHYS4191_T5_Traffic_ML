```python
import pandas as pd
import numpy as np
```

```python
March_05 = pd.read_csv('June_04_FullData.csv')
March_12 = pd.read_csv('June_11_FullData.csv')
March_19 = pd.read_csv('June_18_FullData.csv')
March_26 = pd.read_csv('June_25_FullData.csv')
```

```python
March_05.head()
```

```python
# Check column names and drop the first column ('Unnamed: 0').

March_05.columns
```

```python
March_05.drop(columns=['Unnamed: 0'],inplace=True)
March_12.drop(columns=['Unnamed: 0'],inplace=True)
March_19.drop(columns=['Unnamed: 0'],inplace=True)
March_26.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
March_05.head()
```

```python
# Only interested in data between 01F0339S and 01F0509S. Therefore, filter out irrelevant data.

March_05 = March_05[March_05["2"].str.contains("339S", na=False)]
March_05 = March_05[March_05["4"].str.contains("509S", na=False)]

March_12 = March_12[March_12["2"].str.contains("339S", na=False)]
March_12 = March_12[March_12["4"].str.contains("509S", na=False)]

March_19 = March_19[March_19["2"].str.contains("339S", na=False)]
March_19 = March_19[March_19["4"].str.contains("509S", na=False)]

March_26 = March_26[March_26["2"].str.contains("339S", na=False)]
March_26 = March_26[March_26["4"].str.contains("509S", na=False)]
```

```python
total_data_rows = March_05.shape[0]+March_12.shape[0]+March_19.shape[0]+March_26.shape[0]
print('Total Rows of Data: '+str(total_data_rows))
```

```python
# Combine into one dataframe

df = pd.concat([March_05,March_12,March_19,March_26], ignore_index=True)
df.shape
```

```python
# Save the cleaned raw data file as a csv file.

df.to_csv('Cleaned_Raw_Data.csv')
```

```python
import re

example = df['7']
example
```

```python
# Calculate the latencies and time for each row and append it to the array to use for plotting.

latency = []
time_of_the_day = []

for i in range(len(example)):
    temp = re.split('[+]', example[i])
    t1 = temp[0]
    t2 = re.split('[;]', temp[4])[1]
    t1 = pd.to_datetime(t1)
    t2 = pd.to_datetime(t2)
    diff = pd.Timedelta(t2 - t1).total_seconds()
    latency.append(diff)
    time_of_the_day.append(t1)
```

```python
# Check if the lengths of both arrays match to make sure the data is correct. Here, the index of each array represents the latency/time of its counterpart. For example, for index i, the latency is latency[i] and the time is time_of_the_day[i].

print('Length of Latency Array: '+str(len(latency)))
print('Length of Time Array: '+str(len(time_of_the_day)))
```

```python
time_of_the_day[0]
```

```python
date_df = pd.DataFrame({'TimeStamp':time_of_the_day})
```

```python
date_df.head()
```

```python
# date_df['TimeStamp'] = pd.Series([val.time() for val in date_df['TimeStamp']])
time_only = pd.DatetimeIndex(date_df['TimeStamp']).time
```

```python
time_only[0]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.scatter(time_only, latency, marker = 'x', s=0.5)
plt.title('Latency vs. Time of the day with limit 2020 June Thursdays')
plt.xlabel('Time of the day (hh:mm:ss)')
plt.ylabel('Latency (s)')
plt.xlim(xmin=0)
plt.xlim(xmax=86400)
plt.show()
```

```python
# Since the above plot has a lot of noise with latency values greater than a certain threshold (~1500), any latency values greater than 1500 sohuld be removed. 

# Convert the lists into a dataframe to filter out rows which have latency values > 1500.
new_df = pd.DataFrame({'Time':time_only,'Latency':latency})

new_df.drop(new_df[(new_df['Latency']>1500)].index,inplace=True)
```

```python
# Convert the new dataframe with latencies < 1500 into a list for plotting.

new_time = (new_df['Time'].to_list())
new_latency = (new_df['Latency'].to_list())
```

```python
plt.figure(figsize=(15,10))
plt.scatter(new_time, new_latency, marker = 'x', s=0.5)
plt.title('Latency vs. Time of the day with limit 2020 June Thursdays')
plt.xlabel('Time of the day (hh:mm:ss)')
plt.ylabel('Latency (s)')
plt.xlim(xmin=0)
plt.xlim(xmax=86400)
plt.show()
```

```python
# Convert time to time in hours.

time_in_hours = []
for i in range(len(new_time)):
    temp = new_time[i].hour + new_time[i].minute/60
    time_in_hours.append(temp)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(time_in_hours, new_latency, marker = 'x', s=0.5)
plt.title('Latency vs. Time of the day with limit 2020 June Thursdays')
plt.xlabel('Time of the day (hours)')
plt.ylabel('Latency (s)')
plt.xlim(xmin=0)
plt.xlim(xmax=24)
plt.show()
```

```python
# Plot a polynomial regression line to fit the scatter plot.

x = time_in_hours
y = new_latency
# np.ployfit(x, y, N) where x is the x values, y is the y values, and N is the Nth ploynomial regression.
z = np.polyfit(x, y, 30)
p = np.poly1d(z)
xp = np.linspace(0,24)

plt.figure(figsize=(15,7))
plt.scatter(time_in_hours, new_latency, marker = 'x', s=0.5)
plt.plot(xp, p(xp), '-', marker = 'x', c = 'green')
plt.title('Latency vs. Time of the day for 2020 June Thursdays with latency limit at 1500s')
plt.xlabel('Time of the day (hours)')
plt.ylabel('Latency (s)')
plt.xlim(xmin=0)
plt.xlim(xmax=24) 
plt.show()
```
